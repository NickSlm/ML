```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
```

```python
label_names = {0: "airplane", 1: "automobile", 2: "bird", 3: "cat", 4: "deer", 5: "dog", 
              6: "frog", 7: "horse", 8: "ship", 9: "truck"}
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
val_size = x_test.shape[0] // 2

x_train = x_train / 255.
x_val = x_test[:val_size] / 255.
y_val = y_test[:val_size]
x_test = x_test[val_size:] / 255.
y_test = y_test[val_size:]
```

```python
y_train = to_categorical(y_train, 10)
y_val = to_categorical(y_val, 10)
y_test = to_categorical(y_test, 10)
```

```python
y_train.shape
```

```python
def plot_images(images, n_cols):
    n_images = len(images)
    n_rows = n_images // n_cols + 1
    figure = plt.figure(figsize=(2.28, 1.6))
    for i, image in enumerate(images):
        figure.add_subplot(n_rows, n_cols, i + 1)
        plt.imshow(image, cmap="binary")
        plt.axis("off")
    plt.show()
```

```python
plot_images(x_test[:8], n_cols=4)
```

## Denoised Autoencoder

```python
keras.backend.clear_session()
```

```python
encoder = keras.models.Sequential([
    keras.layers.Input(shape=[32,32,3]),
    keras.layers.GaussianNoise(0.1),
    keras.layers.Conv2D(filters=16, kernel_size=5, padding="same", activation="selu"),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Conv2D(filters=32, kernel_size=3, padding="same", activation="selu"),
    keras.layers.MaxPool2D(pool_size=(2,2)),
])

decoder = keras.models.Sequential([
    keras.layers.Input(shape=[8,8,32]),
    keras.layers.Conv2DTranspose(16, kernel_size=5, strides=(2, 2), padding='same', activation="selu"),
    keras.layers.Conv2DTranspose(3, kernel_size=5, strides=(2, 2), padding='same', activation="sigmoid"),
])
```

```python
autoencoder = keras.models.Sequential([encoder, decoder])
autoencoder.compile(loss="binary_crossentropy", optimizer=keras.optimizers.Nadam(), metrics=["mse"])
```

```python
history = autoencoder.fit(x_train, x_train, epochs=10, validation_data=(x_val, x_val))
```

```python
n_images = x_test[:8]
n_images_noisy = n_images + np.random.randn(8, 32,32,3) * 0.1
n_images_denoised = autoencoder.predict(n_images_noisy)

figure = plt.figure(figsize=(3, 8))
for i in range(8):
    # original image
    figure.add_subplot(8, 3, i * 3 + 1)
    plt.imshow(n_images[i])
    plt.axis("off")
    # denoised image
    figure.add_subplot(8, 3, i * 3 + 2)
    plt.imshow(n_images_noisy[i])
    plt.axis("off")
    # reconstructed image
    figure.add_subplot(8, 3, i * 3 + 3)
    plt.imshow(n_images_denoised[i])
    plt.axis("off")
```

```python
autoencoder.save("my_autoencoder.h5")
```

```python
def plot_acc_loss(history, n_epochs=32):
    epochs = np.arange(n_epochs)

    # Accuracy plot
    figure = plt.figure(figsize=(4.8,2.4))
    figure.add_subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="accuracy")
    plt.plot(history.history["val_accuracy"], label="val_accuracy")
    plt.legend()
    # Loss plot
    figure.add_subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="loss")
    plt.plot(history.history["val_loss"], label="val_loss")
    plt.legend()
    plt.show()
```

## Basic CNN cifar10 Model

#### Version 2.0

```python
keras.backend.clear_session()
```

```python
cnn_classifier = keras.models.Sequential([
    keras.layers.Input(shape=[32,32,3]),
    keras.layers.Conv2D(filters=32, kernel_size=3, padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Dropout(0.25),
    keras.layers.Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Dropout(0.25),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(10, activation="softmax")
])

early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)

cnn_classifier.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(), metrics=["accuracy"])
b_v_2 = cnn_classifier.fit(x_train, y_train, epochs=50, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
plot_acc_loss(b_v_2, 50)
cnn_classifier.evaluate(x_test, y_test)
```

## Pretrained CNN cifar10 Model

```python
keras.backend.clear_session()
```

```python
my_autoencoder = keras.models.load_model("my_autoencoder.h5")
```

```python
encoder, _ = my_autoencoder.layers
encoder.summary()
```

```python
model = keras.models.Sequential(encoder)
model.add(keras.layers.Conv2D(filters=32, kernel_size=3, padding="same", activation="relu"))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.MaxPool2D(pool_size=2))
model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.MaxPool2D(pool_size=2))
model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation="relu"))
model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
for layer in model.layers[0].layers:
    layer.trainable = False
```

```python

early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)

model.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(), metrics=["accuracy"])
p_v_2 = model.fit(x_train, y_train, epochs=50, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
for layer in model.layers[0].layers:
    layer.trainable = True

early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)

model.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.Adam(), metrics=["accuracy"])
p_v_2 = model.fit(x_train, y_train, epochs=50, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
plot_acc_loss(p_v_2, 50)
model.evaluate(x_test, y_test)
```

### Differences between pretrained and not

```python
figure = plt.figure(figsize=(8.6,6.4))

figure.add_subplot(1, 2, 1)
plt.title("Reg")
plt.plot(b_v_2.history["accuracy"], label="accuracy")
plt.plot(b_v_2.history["val_accuracy"], label="val_acc")
plt.legend()
figure.add_subplot(1, 2, 2)
plt.title("Pretrained")
plt.plot(p_v_2.history["accuracy"], label="accuracy")
plt.plot(p_v_2.history["val_accuracy"], label="val_acc")
plt.legend()
```
