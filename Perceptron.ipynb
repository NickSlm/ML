```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
iris = load_iris()
X = iris.data[:,(2,3)]
y = (iris.target == 0).astype(np.int32)
y
```

```python
per_clf = Perceptron(random_state=42)
per_clf.fit(X, y)
print(per_clf.coef_)
print(per_clf.intercept_)
```

```python
a = -per_clf.coef_[0][0] / per_clf.coef_[0][1]  # slope
b = -per_clf.intercept_[0] / per_clf.coef_[0][1]  # intercept
# line formula = y = ax + b
point_1 = [0, a * 0 + b]
point_2 = [5, a * 5 + b]
```

```python
plt.figure(figsize=(6,3))
plt.axis([0,5,0,2])
plt.grid()
plt.scatter(X[:,0], X[:, 1], c=y)
plt.plot([point_1[0], point_2[0]], [point_1[1], point_2[1]])
plt.show()
```
